Here i have created another representation of the model variable with barplot that seems better than the pie chart:
1)At first creating dataset -


```python

df_modelofcars = pd.DataFrame(df["model"].value_counts())
df_modelofcars = df_modelofCars.reset_index(0)
df_modelofcars = df_modelofcars.rename(columns={"model":"no_of_cars"})
df_modelofcars["% of cars"] = (df_modelofcars["no_of_cars"]/df.shape[0])*100
df_modelofcars = df_modelofcars.round(2)

display(df_modelofcars)
```

2)Creating the barplot

```python
sns.barplot(x="model_of_Cars" , 
            y="% of cars", 
            data=df_modelofcars, 
            color="blue",
            alpha=0.8)

plt.xlabel("Car's model Type",color="blue")
plt.ylabel("% of cars",color="blue")
plt.title("Percentage of different models of ford cars",color="Green")
plt.xticks(rotation="vertical")
plt.yticks(np.arange(0,38,2))


plt.show()
```

```python
plt.figure(figsize=(20,15))

plt.subplot(2, 2, 1)
sns.scatterplot(x="mileage", y="price", data=df,hue="fuelType")

plt.subplot(2, 2, 2)
sns.scatterplot(x="mpg", y="price", data=df,hue="fuelType")

plt.subplot(2, 2, 3)
sns.scatterplot(x="engineSize", y="price", data=df,hue="fuelType")

plt.tight_layout()
plt.show()
```

# Regression plot with numerical columns:
    * mileage vs price
    * mpg vs price
    * engineSize vs price

```python


plt.figure(figsize=(20,15))

plt.subplot(2, 2, 1)
sns.regplot(x="mileage", y="price", data=df, line_kws={"color":"red"})

plt.subplot(2, 2, 2)
sns.regplot(x="mpg", y="price",data=df,line_kws={"color":"blue"})

plt.subplot(2, 2, 3)
sns.regplot(x="engineSize", y="price", data=df,line_kws={"color":"green"})

plt.tight_layout()
plt.show()
```
